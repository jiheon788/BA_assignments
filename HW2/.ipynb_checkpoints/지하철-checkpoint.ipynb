```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap as Basemap
import matplotlib.lines as mlines
```

```python
pd.set_option("max_rows", 5)
pd.set_option("max_columns", None)
```

```python
# '역' & '(###)'지우기
def stationNameClear(stations):
    station_list = []
    for station in stations['역사명']:
        if station[-1] == ')':        
            station = ''.join(station.split('(')[:-1])
        if station[-1] == '역':
            station = station[:-1]
        station_list.append(station)
        
    stations['역사명']=station_list
    return stations

# 띄어쓰기 제거
def spaceClear(stations):
    station_list = []
    for station in stations['영문역사명']:
        station_list.append(station.replace(" ",""))

    stations['영문역사명']=station_list
    return stations

#데이터 프레임을 딕셔너리로
def df2dict(df):
    if df.shape[1] != 2 :
        return print('only 2 columns')
    
    df_dict = {}
    for key, value in zip(df.iloc[:, 0], df.iloc[:, 1]):
        df_dict[key] = value
        
    return df_dict
```

```python
stations = pd.read_csv('data/input/전국도시철도역사정보표준데이터.csv')
stationNameClear(stations)
spaceClear(stations)

stationDf=stations[['역사명','영문역사명','역위도','역경도']]
stationDf.rename(columns={'역사명':'kor_name','영문역사명':'eng_name','역위도':'latitude','역경도':'longitude'},inplace=True)

stationName_ix = stationDf['eng_name'].values

metroKorEngDict=df2dict(stationDf[['kor_name','eng_name']])
# md
# stationDf['to_eng']=stationDf['to'].map(md)
stationDf
```

```python
metro = pd.read_csv('data/input/metro.csv')
metro
```

```python
metro['from'] = metro['from'].map(metroKorEngDict)
metro['to'] = metro['to'].map(metroKorEngDict)
metro.dropna(inplace=True)
metro
```

```python
metro = metro[(metro['from'].isin(stationName_ix)) & (metro['to'].isin(stationName_ix))]
metro['total'].median()
```

```python
metroDf = metro[metro['total'] >= 87]
metroDf
```

```python
counts = metroDf['from'].append(metroDf.loc[metroDf['from'] != metroDf['to'], 'to']).value_counts()
counts = pd.DataFrame({'eng_name': counts.index, 'total': counts})
counts
```

```python
pos_data = counts.merge(stationDf, on = 'eng_name')
pos_data
```

```python
graph = nx.from_pandas_edgelist(
    metroDf,
    source = 'from',
    target = 'to',
    edge_attr = 'total',
    create_using = nx.DiGraph())

plt.figure(figsize = (20,19))
nx.draw_networkx(graph)
```
