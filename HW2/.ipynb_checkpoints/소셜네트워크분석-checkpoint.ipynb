# 1. Social Network Basics
* Network: 각각 객체들의 상호 연결된 구조 (**복수의 노드들이 연결된 구조**)
* 구성요소
    * Nodes: 고유한 속성을 가지고 있는 행위자
    * Edges: 노드들관의 관계(relation, link)

```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
import warnings; warnings.simplefilter('ignore')
import pandas as pd
```

```python
pd.set_option("max_rows", 5)
pd.set_option("max_columns", None)
```

## 1.1. Symmetric Networks (undirected)
* 무방향 네트워크

```python
G_symmetric = nx.Graph()

G_symmetric.add_edge('Steven',  'Laura')
G_symmetric.add_edge('Steven',  'Marc')
G_symmetric.add_edge('Steven',  'John')
G_symmetric.add_edge('Steven',  'Michelle')
G_symmetric.add_edge('Laura',   'Michelle')
G_symmetric.add_edge('Michelle','Marc')
G_symmetric.add_edge('George',  'John')
G_symmetric.add_edge('George',  'Steven')

print(nx.info(G_symmetric))
```

```python
plt.figure(figsize=(5,5))
nx.draw_networkx(G_symmetric);
```

## 1.2. Asymmetric Networks (directed)
* 방향 네트워크

```python
G_asymmetric = nx.DiGraph()
G_asymmetric.add_edge('A','B')
G_asymmetric.add_edge('A','D')
G_asymmetric.add_edge('C','A')
G_asymmetric.add_edge('D','E')

nx.spring_layout(G_asymmetric)
nx.draw_networkx(G_asymmetric)
```

## 1.3. Weighted Networks
* 가중치 네트워크

```python
G_weighted = nx.Graph()

G_weighted.add_edge('Steven',  'Laura',   weight=25)
G_weighted.add_edge('Steven',  'Marc',    weight=8)
G_weighted.add_edge('Steven',  'John',    weight=11)
G_weighted.add_edge('Steven',  'Michelle',weight=1)
G_weighted.add_edge('Laura',   'Michelle',weight=1)
G_weighted.add_edge('Michelle','Marc',    weight=1)
G_weighted.add_edge('George',  'John',    weight=8)
G_weighted.add_edge('George',  'Steven',  weight=4)

# 가중치 8 초과
elarge = [(u, v) for (u, v, d) in G_weighted.edges(data=True) if d['weight'] > 8]
# 가중치 8 이하
esmall = [(u, v) for (u, v, d) in G_weighted.edges(data=True) if d['weight'] <= 8]

# positions for all nodes
pos = nx.circular_layout(G_weighted)  

# nodes
nx.draw_networkx_nodes(G_weighted, pos, node_size=700)

# edges
nx.draw_networkx_edges(G_weighted, pos, edgelist=elarge,width=6)
nx.draw_networkx_edges(G_weighted,
                       pos,
                       edgelist=esmall,
                       width=6,
                       alpha=0.5,
                       edge_color='b',
                       style='dashed')
# labels
nx.draw_networkx_labels(G_weighted, pos, font_size=20, font_family='sans-serif')

plt.axis('on')
plt.show()
```

# 2. 네트워크 유형 
* 방향성과 가중치에 따른 분류
* 관계의 위상에 따른 분류
* 네트워크 크기에 따른 분류

* 분석 대상에 따른 분류
    * 전체 vs 하위
    * Cluster(클러스터)
        * 유사한 속성을 가지고 있는 노드들의 집합
        * Clustering: 유사성에 따라 클러스터를 분류하는 작업, 노드간 유사성 분석
        * 수학적 알고리즘을 기반으로 계산 및 시각화
        > Clauset-Newman-Moore 알고리즘, Girvan-Newman 알고리즘
        * Ego network: 특정 노드를 중심에 두고, 그 노드와 직접적으로 연결된 노드들 간의 관계를 네트워크로 표현한 것
    * Clique(파당)
        * 네트워크의 하위 네트워크
        * 파당에 포함된 모든 노드들이 상호간에 직접 연결(완전연결)
        * 다른 파당에는 포함 x
        * 3개 이상의 노드를 포함한 것을 파당이라고 함
        > 3-노드 파당, 4-노드 파당, ...., n-노드 파당 

```python
# 군집화 계수 Clustering coefficient
print(nx.clustering(G_symmetric,'Michelle'))
print(nx.clustering(G_symmetric,'Laura'))
print(nx.average_clustering(G_symmetric))
```

# 3. 소셜 네트워크 분석 지표
* 네트워크 수준의 속성 분석
    * Size
    * Density
    * Inclusiveness
 
* 노드 수준의 속성 분석
    * Degree(연결정도)
        * 노드가 가지고 있는 connections의 갯수
        * 고립된 노드는 연결정도가 0
        * 연결 정도의 측정을 통해 해당 노드의 활동성과 영향력 측정 가능
    * Distance(연결거리)
        * 노드간 관계의 강도
        * Weighted network: 상대적인 가중치로 관계 빈도, 친밀성, 지속성, 신뢰성 등 정의 가능
        * 링크 선의 굵기로 표시
        * 최단 연결거리: 노드간 연결이 짧을수록 연결 높아짐, 정보가 얼마나 신속히 전달되는지 가늠하는 지표
    * Eccentricity(Diameter)
        * 가장 먼 거리 = 네트워크 지름
    * Average Distance
        * 평균연결고리: 평균적으로 최소 몇 단계의 링크를 걸쳐서 연결이 가능한지 보여줌
 

```python
print(f"해당 노드의 Degree: {nx.degree(G_symmetric, 'Michelle')}")
print(f"두 요소의 최단 경로: {nx.shortest_path(G_symmetric, 'Michelle', 'John')}")
print(f"두 요소의 최단 경로 거리: {nx.shortest_path_length(G_symmetric, 'Michelle', 'John')}")
print(f"가장 먼 거리: {nx.eccentricity(G_symmetric,'Michelle')}")
print(f"가장 먼 거리: {nx.eccentricity(G_symmetric,'Steven')}")
```

* Breadth-first search

```python
S = nx.bfs_tree(G_symmetric, 'Steven')
nx.draw_networkx(S)
```

# 4. Centrality measures(중심성 지표)
* Centrality measures: 가장 중요한 노드를 찾는 parameters

## 4.1. Degree Centrality (연결정도 중심성)
* Degree 중심
* 한 노드가 얼마나 많은 다른 노드들과 연결 맺고 있는지 측정

```python
nx.degree_centrality(G_symmetric)
```

## 4.2. Eigenvector Centrality (아이겐 벡터 중심성)
* 연결된 다른 노드들의 중요도를 함께 반영
* 친구가 많은 것 말고도 중요한 친구가 많은 지를 보고자 할 때 사용

```python
nx.eigenvector_centrality(G_symmetric)
```

## 4.3. Closeness Centrality (근접 중심성)
* 한 노드가 얼마나 전체 네트워크의 중심에 근접한지 측정

```python
nx.closeness_centrality(G_symmetric)
```

## 4.4 Betweenness Centrality (매개 중심성)
* 한 노드가 다른 노드들 간의 네트워크 형성에 있어서 중개자 역할을 얼마나 수행하는지 측정

```python
nx.betweenness_centrality(G_symmetric)
```

## 4.5. 중심성을 활용한 시각화 (betweenness_centrality)

```python
pos = nx.spring_layout(G_symmetric)
betCent = nx.betweenness_centrality(G_symmetric, normalized=True, endpoints=True)
node_color = [20000.0 * G_symmetric.degree(v) for v in G_symmetric]
node_size =  [v * 10000 for v in betCent.values()]
plt.figure(figsize=(10,10))
nx.draw_networkx(G_symmetric, pos=pos, with_labels=True,
                 node_color=node_color,
                 node_size=node_size )
```

```python
sorted(betCent, key=betCent.get, reverse=True) #[:5]
```
