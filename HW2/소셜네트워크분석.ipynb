```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap as Basemap
import matplotlib.lines as mlines
```

```python
pd.set_option("max_rows", 5)
pd.set_option("max_columns", None)
```

```python
DG = nx.DiGraph()
DG.add_node('A')
DG.add_node('B')
DG.add_node('C')
DG.add_node('D')
DG.add_node('E')
DG.add_node('F')
DG.add_node('G')
DG.add_node('H')

DG.add_edge('A', 'B')
DG.add_edge('B', 'A')
DG.add_edge('C', 'B')
DG.add_edge('D', 'A')
DG.add_edge('E', 'H')
DG.add_edge('H', 'B')
DG.add_edge('F', 'E')
DG.add_edge('F', 'H')
DG.add_edge('G', 'F')
DG.add_edge('G', 'E')
DG.add_edge('C', 'D')
DG.add_edge('G', 'H')

pos=nx.shell_layout(DG)

nx.draw_shell(DG)
nx.draw_networkx_labels(DG, pos, font_size=10)
plt.show()
```

```python
featuringDF = pd.read_csv('data/input/featuring.csv')
featuringDF
```

```python
graph = nx.from_pandas_edgelist(
    featuringDF,
    source = 'from',
    target = 'to',
    edge_attr = None,
    create_using = nx.DiGraph())
plt.figure(figsize = (10,9))
nx.draw_networkx(graph)
```
